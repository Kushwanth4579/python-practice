```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['sv','sev','c','cl','vdc','eng','it','evpr'])
y = np.array([78,78,75,92,93,90,100,92])
plt.grid()
plt.plot(x, y,color='blue',marker='d',ms=10,mfc='red')
plt.show()
plt.xlabel('subjects')
plt.ylabel('marks')
plt.bar(x,y,color='red',width=0.5)
plt.show()
plt.scatter(x,y)
plt.show()
a=['sv','sev','c','cl','vdc','eng','it','evpr']
plt.pie(y,labels=a)
plt.legend()
plt.show()
plt.hist(y)
plt.show()
plt.plot(x,y,'o:r')
plt.show()
plt.barh(x,y,color='y')
plt.show()
```
